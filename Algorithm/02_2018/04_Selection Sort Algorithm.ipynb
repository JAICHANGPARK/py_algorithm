```python
import random
```

```python
def selected_sort(random_list):
    for sel in range(len(random_list) -1):
        min_value = random_list[sel]
        min_index = sel
        #find min vallue
        for step in range(sel+1, len(random_list)):
            print("step", step)
            if min_value > random_list[step]:
                print(random_list[step])
                min_value = random_list[step]
                min_index = step
            
        #swap
        random_list[min_index] = random_list[sel]
        random_list[sel] = min_value
```

```python
if __name__ == '__main__':
    sort_list =[]
    for i in range(100):
        sort_list.append(random.randint(1,10))
    print("<before sort>")
    print(sort_list)
    selected_sort(sort_list)
    print("<After sort>")
    print(sort_list)
```
